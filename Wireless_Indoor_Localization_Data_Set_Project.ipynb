```python
import pandas as pd
from sklearn.model_selection import train_test_split
# functions to split data into training and testing sets
from sklearn.preprocessing import StandardScaler
# used to standardized features
from sklearn.neighbors import KNeighborsClassifier
# learning algoirthm that helps make classificationhs on data neighbors
from sklearn.metrics import accuracy_score, classification_report
# functions that help to evaluate the performance of the classifier


# load the data
file_path = '/content/wifiLocalization.csv'
data = pd.read_csv(file_path, delimiter='\t', header=None)

# rename columns for easier reference
# data set has 7 features that align with Wifi signal strengths and one variable
# that indicates the room
data.columns = ['Wifi 1', 'Wifi 2', 'Wifi 3', 'Wifi 4', 'Wifi 5', 'Wifi 6', 'Wifi 7', 'Room']

# seperate the feature and target variable
# X contains features
X = data[['Wifi 1', 'Wifi 2', 'Wifi 3', 'Wifi 4', 'Wifi 5', 'Wifi 6', 'Wifi 7']]
# Y contains target variable
y = data['Room']

# normalize the data, aka make it scalar (magnitude and no direction)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# split the data into training and testing sets
# 30% of the data is used for testing, whereas the other 70% is used to train
# the model, the random state ensures that the same randomization is used each
# time to result in the same splits of the data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# train the k-NN classifier
knn = KNeighborsClassifier(n_neighbors=3)
# k = 3, so the model will consider the 3 nearest neighbors to classify
# can be chnaged to find optimal numbers of neighbors that give best performance
knn.fit(X_train, y_train)
# trains the kNN classifier using the training data

# predict on the test set
y_pred = knn.predict(X_test)
# uses trained classifier to predict for test set

# evaluate the model
accuracy = accuracy_score(y_test, y_pred)
# computes accuracy on test set
print(f'Accuracy: {accuracy}')
print(classification_report(y_test, y_pred))
# gives classification report of persion, recall, and f1 score for each class

# display the first few rows of the dataframe
data.head()
```
